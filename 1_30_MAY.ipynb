```python
l=[5,7,9,13,32,33,42,54,56,88]
key=5
l.sort()
print(l)

lo=0
u=len(l)-1

while lo<=u:
  mid=(u+lo)//2

  if l[mid]<key:
    lo=mid+1
  elif l[mid]>key:
    u=mid-1
  else:
    print(f"item {key} found at {mid} th index")
    break
else:
  print(f"item {key} not found")
```

```python
class Student():
  def __init__(self,n,r):
    self.name=n
    self.rollno=r

  def display_data(self):
    print("name=",self.name)
    print("rollno=",self.rollno)


class Test(Student):
  def __init__(self,n,r,m):
    Student.__init__(self,n,r) #class name then self
    self.mark=m

  def display_mark(self):
    print("mark=",self.mark)


class Sports():
  def __init__(self,sm):
    self.smark=sm

  def display_smark(self):
    print("smark=",self.smark)

class Grade(Test,Sports):
  def __init__(self,n,r,m,sm):
    Test.__init__(self,n,r,m)
    Sports.__init__(self,sm)

  def calgrade(self):
    tm=self.mark+self.smark
    p=tm/200 *100
    if p>80:
      print("grade A")
    elif p>50:
      print("grade B")
    else:
      print("failed")

g1=Grade("aby",12,200,150)
# g1.get_data()
# g1.get_mark()
# g1.get_smark()
g1.display_data()
g1.display_mark()
g1.display_smark()
g1.calgrade()
```

```python
#polymorphism
class Point():
  def __init__(self,x,y):
    self.x=x
    self.y=y
  def __add__(self,other):
    x=self.x+other.x+10
    y=self.y+other.y+10
    return Point(x,y)
  def __str__(self):
    return f"({self.x},{self.y})"

p1=Point(1,2)
p2=Point(3,4)

print(p1+p2)
```
